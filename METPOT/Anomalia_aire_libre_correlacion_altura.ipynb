###**Linea de tendencia de la anomalía de aire libre**

La misma es $AAL(h)\,=\,0.02835234426566967h+10.924068556387546$<br/>
<br/>Si la anomalía de Aire Libre se escribe formalmente como:<br/>
<br/>$AAL(h)\,=\,g_{obs}\,-\,\gamma\,+\,F$<br/>
<br/>También la podemos expresar en función de la anomalía de bouguer como:<br/>
<br/>$AAL(h)\,=\,AB\,+\,C_{B}$<br/>
<br/>Donde $C_{B}\,=\,2\pi G \rho h$<br/>
<br/>Así: $AAL(h)\,=\,AB\,+\,2\pi G\rho h$<br/>
<br/>La pendiente de la recta de AAL es: $m\,=\,2\pi \rho G\,=\,0.02835234426566967$<br/>
<br/>La densidad del terreno es: $\rho \,=\,\dfrac{0.02835234426566967}{2\pi G}$<br/>
<br/>$\rho \,=\,¿676.118\,kg/m^3?$
